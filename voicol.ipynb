```python
import numpy as np
import struct
import pyaudio as pa
import wave
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import librosa as rosa
```

```python
CHUNK = 1024*2
FORMAT=pa.paInt16
CHANNELS=1
RATE = 44100
```

```python

p=pa.PyAudio()
stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK,
)



fig, ax = plt.subplots()
x = np.arange(0, 2 * CHUNK, 2)
(line,) = ax.plot(x, np.random.rand(CHUNK), "r")
ax.set_ylim(-32770, 32770)
ax.set_xlim = (0, CHUNK)
fig.show()


while 1:
    data = stream.read(CHUNK)
    dataInt = struct.unpack(str(CHUNK) + "h", data)
    line.set_ydata(dataInt)  
    fig.canvas.draw()
    # plt.show()
    fig.canvas.flush_events()
```
